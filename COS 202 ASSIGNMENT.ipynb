```python
print("--- MATHEMATICAL CALCULATOR ---")
print("Operators: +, -, *, /, \\ (integer div), ^ (power), % (modulus)")
print("Type 'C' to clear/restart, or 'OFF' to quit.")
print("-------------------------------------")

while True:
    num1_input = input("\nEnter first number (or 'OFF' to quit): ")
    print("Your input:", num1_input)
    if num1_input.upper() == 'OFF':
        print("Turning off... Goodbye!")
        break
    try:
        num1 = float(num1_input)
    except ValueError:
        print()
        continue
    op = input("Enter operator (+, -, *, /, \\, ^, %): ")
    print("Your input:", op)
    
    
    if op.upper() == 'OFF':
        print("Turning off... Goodbye!")
        break
    if op.upper() == 'C':
        print("Cleared! Starting fresh.")
        continue
    num2_input = input("Enter second number: ")
    print("Your input:", num2_input)
    if num2_input.upper() == 'OFF':
        print("Turning off... Goodbye!")
        break
    if num2_input.upper() == 'C':
        print("Cleared! Starting fresh.")
        continue
        
    try:
        num2 = float(num2_input)
    except ValueError:
        print()
        continue

    if op == '+':
        result = num1 + num2
    elif op == '-':
        result = num1 - num2
    elif op == '*':
        result = num1 * num2
    elif op == '/':
        if num2 == 0:
            result = "Error: Division by zero!"
        else:
            result = num1 / num2
    elif op == '\\':
        if num2 == 0:
            result = "Error: Division by zero!"
        else:
            result = num1 // num2  # Double slash is Python's floor division
    elif op == '^':
        result = num1 ** num2      # Double asterisk is Python's exponent power
    elif op == '%':
        if num2 == 0:
            result = "Error: Division by zero!"
        else:
            result = num1 % num2   # Percent sign finds the remainder
    else:
        result = "Unknown operator!"
    print("Result:" ,result)

```

```python
print("=========================================")
print("   PERSONAL POCKET CGPA CALCULATOR  ")
print("=========================================")
print("Carry your academic records in your pocket!")
print("Type 'QUIT' to exit.\n")

total_grade_points = 0.0
total_credit_units = 0.0
semester_count = 0

while True:
    semester_count += 1
    print("--- SEMESTER" ,semester_count ,"---")
    
    action = input("Press Enter to add this semester (or type 'QUIT' to finish): ")
    if action.strip().upper() == 'QUIT':
        break
        
    sem_grade_points = 0.0
    sem_credit_units = 0.0
    course_count = 0
    
    while True:
        course_count += 1
        print("\nCourse" ,course_count,":")
        
        units_input = input("Enter Credit Units (e.g., 2, 3, 4) or 'DONE' to finish semester: ")
        if units_input.strip().upper() == 'DONE':
            break
            
        try:
            units = float(units_input)
            if units <= 0:
                print("Credit units must be greater than zero. Try again.")
                course_count -= 1
                continue
        except ValueError:
            print("Invalid input. Please enter a valid number for credit units.")
            course_count -= 1
            continue
        
            
        grade = input("Enter Letter Grade (A, B, C, D, E, F): ").strip().upper()
        
        if grade == 'A':
            weight = 5.0
        elif grade == 'B':
            weight = 4.0
        elif grade == 'C':
            weight = 3.0
        elif grade == 'D':
            weight = 2.0
        elif grade == 'E':
            weight = 1.0
        elif grade == 'F':
            weight = 0.0
        else:
            print("Invalid grade entered! Course skipped. Please use A, B, C, D, E, or F.")
            course_count -= 1
            continue
            
        sem_credit_units += units
        sem_grade_points += (units * weight)

    if sem_credit_units > 0:
        sem_gpa = sem_grade_points / sem_credit_units
        print("\n>> Semester:", semester_count,"Summary <<")
        print("Total Units: ",sem_credit_units)
        print("Semester GPA: ",sem_gpa)
        
        total_credit_units += sem_credit_units
        total_grade_points += sem_grade_points
    else:
        print("\nNo valid courses recorded for Semester",semester_count,".")
        semester_count -= 1 

print("\n=========================================")
print("         FINAL PPC CALCULATIONS      ")
print("=========================================")

if total_credit_units > 0:
    cgpa = total_grade_points / total_credit_units
    print("Total Semesters Analyzed:", semester_count)
    print("Total Cumulative Units:",total_credit_units)
    print("Your Overall CGPA:",cgpa)
    
    print("Academic Standing: ", end="")
    if cgpa >= 4.50:
        print("First Class Honors! ")
    elif cgpa >= 3.50:
        print("Second Class Upper  ")
    elif cgpa >= 2.40:
        print("Second Class Lower   ")
    elif cgpa >= 1.50:
        print("Third Class Graduate")
    elif cgpa >= 1.00:
        print("Pass")
    else:
        print("Fail")
else:
    print("No data recorded. The calculator has closed.")
print("=========================================")
```
