```python

print("A code to figure if your digits are increasing, decreasing or equal")
num= int(input("Enter your three digit integer: "))
print("The number you entered is: ",num)
# checking the digits of the number
check1=num//100
check2=num//10
check2=check2%10
check3=num%10


if check1<check2<check3:
    print("The digits are strictly increasing")
elif check1>check2>check3:
    print("The digitd are strictly decreasing")
elif check1==check2==check3:
    print("All the digits are equal")
else:
    print("The digits are neither increasing nor decreasing")
    
```

```python

balance=75000
print("Welcome to your ATM machine")
print("Balance:N",balance)
print()
print("==========MENU==========")
print("1. Deposit Money")
print("2. Withdraw Money")
print("3. Transfer Money")
print("4. Balance Enquiry")
print("5. Exit")

choice=int(input("Enter your choice of transaction(1-5)"))
match choice:
    case 1:
        deposit=int(input("Enter the amount you want to deposit:"))
        if deposit>0:
            balance=balance+deposit
            print("You have successfully deposited:N",deposit)
            print("Your new balance is:N",balance)
        else:
            print("Deposit amount must be greater than zero")
    case 2:
        withdraw=int(input("Enter your withdrawal amount:"))
        if withdraw>balance:
            print("Insufficient funds.Your available balance is:N",balance)
        else:
            balance=balance-withdraw
            print("You have successfully withdrawn:N",withdraw)
            print("Your new balance is:N",balance)
    case 3:
        recipient_account=input("Enter the recipient's account number:")
        trs_amount=int(input("Enter the amount you want to transfer:"))
        if trs_amount>balance:
            print("Insufficient funds.Your available balance is:N",balance)
        else:
            balance=balance-trs_amount
            print("You have successfully transferred:N",trs_amount,"to account number:",recipient_account)
            print("Your new balance is:N",balance)
    case 4:
        print("Your available balance is:N",balance)
    case 5:
        print("Thank you for using our ATM. Goodbye!")
    case _:
        print("Invalid choice. Please select a valid option from the menu(1-5).")   
```

```python


print("Triangle type checker")
a=float(input("Enter the length of the first side;"))
b=float(input("Enter the length of the second side;"))
c=float(input("Enter the length of the third side;"))

if a==b and b==c and a==c:
    print("The triangle is an Equilateral triangle")
elif a==b or b==c or a==c:
    print("The triangle is an Isosceles triangle")
elif a!=b and b!=c and a!=c:
    print("The triangle is a Scalene triangle")
```

```python
print("Student category determiner")
score1=int(input("Enter your first exam score:"))
score2=int(input("Enter your second exam score:"))
score3=int(input("Enter your third exam score:"))

avg=float((score1+score2+score3)/3)
print("Your average is:",avg)

if avg>=80:
    print("Excellent Student")
elif 80>avg>=60:
    print("Good Student")
elif 60>avg>=50:
    print("Average Student")
else:
    print("Probation Student")
```
